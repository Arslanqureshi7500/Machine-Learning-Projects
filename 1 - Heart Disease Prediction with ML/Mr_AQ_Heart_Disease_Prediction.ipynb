# **OBJECTIVE 2: MACHINE LEARNING**
Next, I will feed these features into various classification algorithms to determine the best performance using a simple framework: Split, Fit, Predict, Score It.

# **Target Variable Splitting**
We will spilt the Full dataset into **Input** and **target** variables

Input is also called **Feature Variables** Output referes to **Target variables**

```python
#use x and y variable to split the training data into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .30, random_state=21)
```

```python
#shape of x_train
print("The x_train feature rows and columns respectively: ", x_train.shape)
x_train.head()
```

```python
#shape of y_train
print("The y_train feature rows and columns respectively: ", y_train.shape)
y_train.head()
```

```python
#shape of x_test
print("The x_test feature rows and columns respectively: ", x_test.shape)
x_test.head()
```

```python
#shape of y_test
print("The y_test feature rows and columns respectively: ", y_test.shape)
y_test.head()
```

# **Logistic Regression**

## **Model Training**

```python
#import model
from sklearn.linear_model import LogisticRegression
```

```python
print("Logistic Regression")

#create instance of model
log_reg = LogisticRegression()

## Pass training data into model for fit
log_reg.fit(x_train, y_train)
```

```python
#scor of model
socre = log_reg.score(x_train, y_train)*100

print('The Logistic Regression Score: ', socre)
```

# **Model Evaluation**

**Logistic Regression**

```python
from sklearn.metrics import accuracy_score
# prediction from the model
y_pred_log_reg = log_reg.predict(x_test)
# Score It

print('Logistic Regression')
# Accuracy
print('--'*30)
log_reg_accuracy = round(accuracy_score(y_test, y_pred_log_reg) * 100,2)
print('Accuracy', log_reg_accuracy,'%')
```

# **Decision Tree**

## **Model Training**

```python
from sklearn.tree import DecisionTreeClassifier

print('Decision Tree Classifier')

#create instance of model
Dtree = DecisionTreeClassifier()

#Pass training data into model
Dtree.fit(x_train, y_train)
```

# **Model Evaluation**
## **Decision Tree Classifer**

```python
from sklearn.metrics import accuracy_score

#predict from the model
y_predict_Decision_tree = Dtree.predict(x_test)

#Score it
print("Decision Tree Classifier")
#Accuracy
Dtree_accuracy = round(accuracy_score(y_predict_Decision_tree, y_test) * 100, 2)
print('Accuracy: ', Dtree_accuracy,'%')
```

# **Random Forest**
## **Model Training**

```python
from sklearn.ensemble import RandomForestClassifier

print("Random Forest Classifier")

#Create instance  of model
rfc = RandomForestClassifier()

#Pass training data into model
rfc.fit(x_train, y_train)
```

# **Model Evaluation**

## **Random Forest**

```python
from sklearn.metrics import accuracy_score

y_predict_rfc = rfc.predict(x_test)

print("Random Forest Classifier")
#Accuracy
rfc_accuracy_score = round(accuracy_score(y_predict_rfc, y_test)*100, 2)
print("Accuracy: ", rfc_accuracy_score,"%")
```

# **Gradient Boosting Classifier**

## **Model Training**

```python
from sklearn.ensemble import GradientBoostingClassifier

print('Gradient Boosting Classifier')
#Create instance of model
gbc = GradientBoostingClassifier()

#Pass training data into mode
gbc.fit(x_train, y_train)
```

# **Model Evaluation**

## **Gradient Boosting Classifier**

```python
from sklearn.metrics import accuracy_score

y_predict_gbc = gbc.predict(x_test)

print("Gradient Boosting Classifier")
print('--'*30)
#Accuracy
gbc_accuracy = round(accuracy_score(y_predict_gbc, y_test)* 100, 2)
print("Accuracy: ",gbc_accuracy,'%')
```

```python
#Sample model score

model_scores = {
    "Logistic Regression":log_reg_accuracy,
    "Decision Tree Classifier": Dtree_accuracy,
    "Random Forest Classifer":rfc_accuracy_score,
    "Gradient Boosting Classifer" :gbc_accuracy
}

# Sort the model scores in descending order based on their values (higher values first)
sorted_scores = sorted(model_scores.items(), key=lambda x: x[1], reverse=True)

# Display the ranking of the models
print("Model Rankings (Greater Values are better):")
for rank, (model_name, score) in enumerate(sorted_scores, start=1):
    print(f"{rank}. {model_name}: {score}")
```
